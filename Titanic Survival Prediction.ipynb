```python
columns_to_remove = ['PassengerId', 'Name', 'Ticket', 'Cabin']
```

```python
df.drop(columns=columns_to_remove, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.impute import SimpleImputer
si_embarked = SimpleImputer(strategy='most_frequent')
si_age = SimpleImputer()

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
np.isnan(X_train_age)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe_embarked = OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe_sex = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)
X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])

X_test_embarked = ohe_embarked.transform(X_test_embarked)
X_test_sex = ohe_sex.transform(X_test[['Sex']])
```

```python
X_train_embarked
```

```python
X_train_sex
```

```python
from sklearn.preprocessing import StandardScaler
sc_age = StandardScaler()
sc_fare = StandardScaler()

sc_age.fit(X_train[['Age']])
sc_fare.fit(X_train[['Fare']])

X_train_age = sc_age.transform(X_train_age)
X_train_fare = sc_fare.transform(X_train[['Fare']])

X_test_age = sc_age.transform(X_test_age)
X_test_fare = sc_fare.transform(X_test[['Fare']])
```

```python
X_train_rem = X_train.drop(columns=['Sex', 'Age', 'Fare', 'Embarked'])
X_test_rem = X_test.drop(columns=['Sex', 'Age', 'Fare', 'Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem, X_train_sex, X_train_embarked, X_train_age, X_train_fare), axis=1)
X_test_transformed = np.concatenate((X_test_rem, X_test_sex, X_test_embarked, X_test_age, X_test_fare), axis=1)
```

```python
X_train_transformed
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_transformed, y_train)
```

```python
 y_pred = lr.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train_transformed, y_train)
```

```python
 y_pred = dtc.predict(X_test_transformed)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_transformed, y_train)
```

```python
 y_pred = rfc.predict(X_test_transformed)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_transformed, y_train)
```

```python
y_pred = svc.predict(X_test_transformed)
```
